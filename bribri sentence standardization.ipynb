(Función para hacer los cambios).

```
def convertToStandard(input, source):
  
  # x = nasal
  # q = lax vowel (ö=oq, ë=eq)

  output = input

  # Separate non-orthographic signs from words
  for c in nonLanguageSigns: output = output.replace(c, nonLanguageSigns.get(c))

  # Normalize combining characters
  for c in normalizeCombiners: output = output.replace(c, normalizeCombiners.get(c))

  # Normalize nasals combiners
  for c in normalizeNasals: output = output.replace(c, normalizeNasals.get(c))

  # Transform complex nasals into Vx
  for c in complexNasalsToX: output = output.replace(c, complexNasalsToX.get(c))
  
  # Transform lax umlauts into "q"
  for c in complexUmlauts: output = output.replace(c, complexUmlauts.get(c))

  # Words that need to be exchanged
  
  almostFinalOutput = ""
  periodAtEnd = ""
  outputTokens = output.split()
  
  for t in outputTokens:
    
    if (t[-1:] == ".") or (t[-1:] == "!") or t[-1:] == "?":
      periodAtEnd = " " + t[-1:]
      t = t[:-1]
    
    for w in exchangeableWords:
      if t == w:
        t = t.replace(w, exchangeableWords.get(w))
        
    if ("GrammarJara 2018" in source):
      for w in exchangeableJaraWords:
        if t == w:
          t = t.replace(w, exchangeableJaraWords.get(w))
    
    for r in exchangeableRoots: 
      if r in t:
        t = t.replace(r, exchangeableRoots[r])
  
    almostFinalOutput += t  + periodAtEnd + " "
    periodAtEnd = ""
  
  output = almostFinalOutput[:-1]

  # Final details
  for c in finalDetails: output = output.replace(c, finalDetails.get(c))
  for i in range(0,8): output = output.replace("  "," ")

  if ("GrammarJara 2018" in source):
    for c in grammarJaraNasals: output = output.replace(c, grammarJaraNasals.get(c))
    print(output)

  return output

#input = "Ye' shkè̠na̠ bua'ë . | Ye' dör bikâkala . | Ìs a' shkè̠na̠ ? | Sa' shkè̠na̠ bua'ë . | Ìs be' shkè̠na̠ ? | Ye' shkè̠na̠ bua'ë . | Ye' dör Juan . | Í̠ma̠ be' kie ? | Ye' kie Juan . | À̠a̠ !"
input = "Ye' kie Hernán , enã be' ? | ajkö́nu̱k, ajkö́no̱ ie | ajkö̀weìnu̱k a̱ñì̱ ta̱, kë̀ e' ta̱', kë̀ e' ta̱' | pë' a̱ñì̱ dalë́rke̱ buâëme̱, a̱ñì̱ dalë́rtsù̱ke̱ buâëme̱ | se' tulúr bë̂rë | bë̂rë se' tulúr, ma̱ éjkë sö e' chè dè"
#input = "be' mí̠ke̠ i sa̠ú̠k | i ta̠wé̠ yö | i sú̠ bö"
#input = "ãwì̱ | ¿be' | wĩ̂wĩ̂ | u̱léx | bë̂rë se' tulúr | sũ̂mĩ! | ùrikĩ? | ieʼrö. | ye' ë́l yë̂ar i tàyë â̠mi̠k."
#input = "à̠ritsök | uyë̀jkuö | ayë́jkuö õ̀ritsök | õ̀ritsè | Ta̠î̠ | tã́ĩ | wĩ̂wĩ̂"
#input = "yéqblök' | wámblë' | balo' | digo \"hola\" | Ye' r i-sau | blo' | kṍ | kã́ | ñõlõ̀ | ñãlã̀ | kõ̀lĩ | kã̀lĩ | pà | apà | kö̀ | jkö̀ | ajkö̀ | kö́ | ajkö́ | krò | dakarò | "
#input = "õ'mĩk | îxñẽ | ĩnúxköl | ĩnúxköl | kãnèxbalök | Si̠ni̠' akayàwa dör darë̂rë̂"
#input = "Nãmùx tö i che: \"¡Ëkèqke!\" | Ềkèqke | a̠mì̠ | a̠ñì̠ | añĩ̀es | a̠ñì̠es. | ye’ tö i õ̀rtsèke! | nãmã̀? | mè̠sa | kṍwötã | ñĩ̀"
#input = "kóxwötã' | Bikàkla | kṍwötã | Kṍbita | sulùë e' kũẽ́k. | Kë́këpa kula' | Lö̀nekã tãĩ̀, ema erë ie' i mowẽ̀ne. | Ie' kĩ i kiàne i mòsopa iã katànok, yànok. | E' kũẽ́ki e' tö e' ẽ̀wẽwã, ema e' bùkàmi di' ã.  | E' ské tskìne i di, e' rö Tchõ'dawe. | Tchõ'dawe bák ie'pa ñìwáblök sulùë."
input = "e'tã e' ã | ie' kĩ tãĩë̀. | tãĩ̀ | tãĩë̀ | Mĩk̀ ã i làla batàla, i lábusi dùnẽ̀kã tã | E' kũẽ́k kë́këpa bák iõ̀iõ̀ tã  | kũẽ́k | nõ̀ũyöla | nõ̀ũsila"
print(input)
print(convertToStandard(input, "bribri"))
print(convertToStandard(input, "GrammarJara 2018 pg226"))
```
